## Filtering Dataframes by Many conditions

### Import de bibliotecas

```python
import pandas as pd
```

```python
titanic = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/titanic.csv')
```

```python
mask1 = titanic.sex == 'male'
```

```python
mask2 = titanic.age > 14
```

```python
(mask1 & mask2).head()
```

```python
male_survived = titanic.loc[mask1 & mask2, ["survived", "pclass", "sex", "age"]]
male_survived
```
