### Filtering DataFrames with one Condition

#### Import de bibliotecas

```python
import pandas as pd
```

```python
titanic = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/titanic.csv')
```

```python
titanic.head()
```

```python
titanic.sex.head()
```

```python
titanic.sex == 'male'
```

```python
titanic[titanic.sex == 'male']
```

```python
mask = titanic.sex == "male"
mask
```

```python
titanic.loc[titanic.sex == "male"]
```

```python
titanic_male = titanic.loc[mask]
```

```python
titanic_male
```

#### filtering by data types

```python
titanic_male.dtypes
```

```python
titanic_male.dtypes == 'object'
```

```python
mask2 = titanic_male.dtypes == 'object'
titanic_obj = titanic.loc[:, mask2]
```

```python
titanic_obj
```

#### oposite mask "~"

```python
titanic_obj2 = titanic.loc[:, ~mask2]
```

```python
titanic_obj2
```

#### Filtering row and columns

```python
titanic.loc[mask, ~mask2]
```
