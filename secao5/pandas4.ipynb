## Advanced Filtering with between(), isin() and ~

### Import de Bibliotecas

```python
import pandas as pd
```

```python
summer = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/summer.csv')
```

```python
summer.head()
```

```python
mask1 = summer.Year == 1988
```

```python
og_1988 = summer.loc[mask1]
```

```python
og_1988.head()
```

```python
og_1988.info()
```

```python
mask2 = summer.Year >= 1992
```

```python
og_since_1992 = summer.loc[mask2]
og_since_1992
```

```python
mask3 = summer.Year.between(1960, 1969, inclusive=True)
```

```python
og_60s = summer.loc[mask3]
og_60s
```

```python
my_favourite_games = [1972, 1996]
```

```python
mask4 = summer.Year.isin(my_favourite_games)
```

```python
summer.loc[mask4]
```

```python
og_not = summer.loc[~mask4]
og_not
```

```python
og_not.Year.unique()
```
