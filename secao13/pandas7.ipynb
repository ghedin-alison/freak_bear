# Transformation with transform()

```python
import pandas as pd
```

```python
tit = pd.read_csv("../Course_Materials_Part2/Video_Lecture_NBs/titanic.csv")
```

```python
tit.head()
```

```python
tit.groupby(["sex", "pclass"])[["survived"]].mean()
```

```python
tit["group_surv_rate"] = tit.groupby(["sex", "pclass"])[["survived"]].transform("mean")
```

```python
tit
```

```python
tit["outliers"] = abs(tit.survived-tit.group_surv_rate)
```

```python
tit
```

```python
# filtrar as pessoas que são exceção em indice dentro do agrupamento.
# se o indice de sobrevivencia do grupo é grande e a pessoa morreu, ele é um outlier
tit[tit.outliers > 0.85]
```
