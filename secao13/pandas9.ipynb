# Generalizing split-apply-combine with aplly()

```python
import pandas as pd
```

```python
tit = pd.read_csv("../Course_Materials_Part2/Video_Lecture_NBs/titanic.csv", usecols=["survived",
                                                                                      "pclass",
                                                                                      "sex",
                                                                                      "age",
                                                                                      "fare"
                                                                                      ])
```

```python
tit.head()
```

```python
tit.groupby("sex").mean()
```

```python
female_group = list(tit.groupby("sex"))[0][1]
```

```python
female_group
```

```python
female_group.mean()
```

```python
def group_mean(group):
    return group.mean()
```

```python
group_mean(female_group)
```

```python
tit.groupby("sex").apply(group_mean)
```

```python
tit.nlargest(5, "age")
```

```python
def five_oldest_surv(group):
    return group[group.survived == 1].nlargest(5, "age")
```

```python
tit.groupby("sex").apply(five_oldest_surv)
```
