# Advanced Aggregation with agg()

```python
import pandas as pd
```

```python
tit = pd.read_csv("../Course_Materials_Part2/Video_Lecture_NBs/titanic.csv", 
                  usecols=["survived", "pclass", "sex", "age", "fare"])
```

```python
tit.head()
```

```python
tit.groupby("sex").mean()
```

```python
tit.groupby("sex").sum()
```

```python
tit.groupby("sex").agg(["mean", "sum", "min", "max"])
```

```python
tit.groupby("sex").agg({"survived": ["mean", "sum"], 
                        "pclass":"mean", 
                        "age": ["mean", "median"],
                        "fare": "max"})
```
