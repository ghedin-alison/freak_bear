# Replacing Na Values by group-specific Values

```python
import pandas as pd
```

```python
tit = pd.read_csv("../Course_Materials_Part2/Video_Lecture_NBs/titanic.csv")
```

```python
tit.head(20)
```

```python
tit.info()
```

```python
mean_age = tit.age.mean()
mean_age
```

```python
tit.age.fillna(mean_age)
```

```python
tit.groupby(["sex", "pclass"])[["age"]].mean()
```

## Função muito utilizada

```python
tit["group_mean_age"] = tit.groupby(["sex", "pclass"])[["age"]].transform("mean")
```

```python
tit[["group_mean_age"]]
```

```python
tit.head(20)
```

```python
tit.age.fillna(tit.group_mean_age, inplace=True)
```
