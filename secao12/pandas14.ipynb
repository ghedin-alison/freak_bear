# pd.merge and join()

```python
import pandas as pd
```

```python
men2004 = pd.read_csv("../Course_Materials_Part2/Video_Lecture_NBs/men2004.csv", index_col="Athlete")
men2008 = pd.read_csv("../Course_Materials_Part2/Video_Lecture_NBs/men2008.csv", index_col="Athlete")
```

```python
men2004.head()
```

```python
men2008.head()
```

```python
men2004.merge(men2008, how="outer", on="Athlete", suffixes=("_2004", "_2008"), indicator=True)
```

```python
pd.merge(men2004, men2008, how="outer", on="Athlete", suffixes=("_2004", "_2008"), indicator=True)
```

```python
# usar somente quando os índices são iguais
# altera a ordenação
men2004.join(men2008, how="outer", lsuffix="_2004", rsuffix="_2008")
```
