# Joining on different Column Labels & Indexes

```python
import pandas as pd
```

```python
men2004 = pd.read_csv("../Course_Materials_Part2/Video_Lecture_NBs/men2004.csv")
men2008 = pd.read_csv("../Course_Materials_Part2/Video_Lecture_NBs/men2008.csv")
```

```python
men2004.head()
```

```python
men2008.head()
```

```python
men2004.columns = ["Name", "Medals"]
```

```python
men2004.merge(men2008, how="outer", left_on="Name", right_on="Athlete", 
             suffixes=("_2004", "_2008"), indicator=True)
```

```python
men0408 = men2004.merge(men2008, how="outer", left_on="Name", right_on="Athlete", 
             suffixes=("_2004", "_2008"), indicator=True)
```

```python
men0408
```

```python
men0408.Name.fillna(men0408.Athlete, inplace=True)
```

```python
men0408
```

```python
men0408.drop(["Athlete", "_merge"], axis=1, inplace=True)
```

```python
men0408
```

```python
men2008.set_index("Athlete", inplace=True)
```

```python
men0408 = men2004.merge(men2008, how="outer", left_on="Name", right_index=True,
             suffixes=["_2004", "_2008"], indicator=True).reset_index()
```
