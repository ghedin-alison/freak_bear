# Hierarchical Indexing(Slicing)

### Import de bibliotecas

```python
import pandas as pd
```

```python
tit = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/titanic.csv')
```

```python
tit = tit.iloc[:50,]
```

```python
tit
```

```python
tit = tit.set_index(["pclass", "sex"]).sort_index(ascending=[True])
```

```python
tit
```

```python
tit.loc[1]
```

```python
tit.loc[[1,2]]
```

```python
tit.loc[:2]
```

```python
tit.loc[1, "female"]
```

```python
tit.loc[1, "female", "age"]
```

```python
tit.loc[(1, "female")]
```

```python
tit.loc[(1, "female"), "age"]
```

```python
tit.loc[([1,2], "female"), ["age", "fare"]]
```

#### Cuidado ao fazer slice de multiindex, se quiser manter todos os dados depois, tem que colocar ":", caso contrrio retorna erro

```python
tit.loc[([1,2], "female"), :]
```

```python
tit.loc[(slice(None), slice("female")), :]
```
