# nunique(), nlargest(), and smallest() with DataFrames

### Import de Bibliotecas

```python
import pandas as pd
```

```python
titanic = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/titanic.csv')
```

```python
titanic.head()
```

```python
titanic.tail()
```

```python
titanic.age.unique()
```

```python
titanic.nunique(axis=0, dropna=False)
```

```python
titanic.nlargest(n = 5, columns= "fare")
```

```python
titanic.sort_values(by="fare", ascending=False).head(5)
```

```python
titanic.nsmallest(n=5, columns="age")
```

```python
titanic.loc[titanic.age.idxmin()]
```
