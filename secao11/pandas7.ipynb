# Detection of Duplicates

```python
import pandas as pd
```

```python
alphabet = pd.DataFrame(["a", "b", "c", "c", "d", "e", "f", "g", "g", "g", "g",], columns = ["Alphabet"])
```

```python
alphabet
```

```python
alphabet.duplicated(keep="first")
```

```python
alphabet[alphabet.duplicated(keep=False)]
```

```python
alphabet.drop_duplicates(keep="first", inplace=True, ignore_index=True)
```

```python
alphabet
```
