## The plot() method

### Import de bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/titanic.csv')
```

```python
titanic.head()
```

```python
titanic.plot(subplots= True, figsize=(25,12), sharex = False)
plt.show()
```

```python
titanic.age.plot(figsize=(25,12))
plt.show()
```

## Customization

```python
titanic.age.plot(figsize=(12,8), 
                 c = 'green', 
                 linestyle='-', 
                 xlim=(0, 900), 
                 ylim=(0,80), 
                 xticks=xticks, 
                 yticks=yticks, 
                 rot = 45)
plt.title("Titanic - Ages", fontsize=15)
plt.legend(loc="best", fontsize=15)
plt.xlabel("Passenger No", fontsize=13)
plt.ylabel("Age", fontsize=13)
```

```python
xticks = [x for x in range(0,901,50)]
```

```python
yticks = [y for y in range(0, 81, 5)]
```
