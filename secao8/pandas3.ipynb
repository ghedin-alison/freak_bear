## Barcharts and Piecharts

### Import de bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
summer = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/summer_2012.csv', index_col="Country")
```

```python
summer
```

```python
summer.Medal.plot(kind = 'bar', figsize=(12,8), fontsize=12, rot=45)
plt.show()
```

```python
summer.Medal.plot(kind='pie', figsize=(12,8), fontsize=12)
plt.show()
```
