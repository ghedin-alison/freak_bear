## Histograms

### Import de bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
tit = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/titanic.csv')
```

```python
tit.head()
```

```python
tit.age.value_counts(bins=10)
```

```python
tit.age.plot(kind = "hist", figsize=(12,8), fontsize=15, bins = 5)
plt.show()
```

#### Using .hist directly to plot a histogram

```python
tit.age.hist(figsize=(12,8), bins=20, xlabelsize=15, ylabelsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.hist(tit.age, bins=80, density=False, cumulative=False)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.hist(tit.age, bins=80, density=False, cumulative=True)
plt.show()
```
