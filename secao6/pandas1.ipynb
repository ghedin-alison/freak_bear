# Best Practices Manipulating elements in a DataFrame

### Import de bibliotecas

```python
import pandas as pd
```

```python
titanic = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/titanic.csv')
```

```python
titanic.head()
```

#### Changing a single value(with loc)

```python
titanic.loc[1, 'age'] = 40
```

```python
titanic.head()
```

#### Changing a single value(with iloc)

```python
titanic.iloc[1, 3] = 41
```

```python
titanic.head()
```

#### Changing multiple values in a column(with loc)

```python
titanic.loc[1:3, 'age'] = 42
```

```python
titanic.head()
```

#### Changing multiple values in a column(with iloc)

```python
titanic.iloc[1:4, 3] = [43, 44, 45]
```

```python
titanic.head()
```

#### Changing multiple values in a column(with boolean indexing)

```python
index_babies = titanic.loc[titanic.age < 1, "age"].index
```

```python
index_babies
```

```python
titanic.loc[titanic.age < 1, 'age'] = 1
```

```python
titanic.loc[index_babies]
```

#### Changing multiple values in a row

```python
titanic.head()
```

```python
titanic.loc[0, "survived": "sex"] = [1, 1, "female"]
```

```python
titanic.head()
```

#### Changing multiples values in multiple rows/columns

```python
titanic.replace(0, "Zero")
```
